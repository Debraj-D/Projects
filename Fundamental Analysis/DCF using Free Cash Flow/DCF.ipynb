```python
import numpy as np

def format_value(val):
    if not pd.isna(val) and val != "nan":
        # Divide by 1e6 and round to 2 decimal places
        return round(val / 1e6, 2)
    return np.NaN
```

```python
# Convert to date-time index and retain only the year
df.index = pd.to_datetime(df.index)
df.index = df.index.year

# Convert all values to millions and round off to 2 decimal places
df = df.map(format_value)
```

```python
# Reverse dataframes rows to have the latest year at the bottom
df = df[::-1]
df
```

```python
# Dropping 2020 as most values are NaN
df.drop([2020],inplace=True)
```

## Assumptions

```python
risk_free_rate = 0.0685     # 10 year India Government Bond Yield
company_growth_rate = 0.20   # According to growth estimate from Yahoo Finance
expected_market_return = 0.12   # Average 15 year NIFTY 50 return
beta = company_info["beta"]
marketcap = company_info["marketCap"]
```

```python
CAPM_discount_rate = risk_free_rate + (expected_market_return - risk_free_rate) * beta
CAPM_discount_rate
```

## Make FCF Projections for next 5 years

```python
last_year = df.iloc[-1, :]
years = range(df.index[-1]+1, df.index[-1] + 6)
FCF_proj = pd.DataFrame(index=years, columns=["Free Cash Flow"])
```

```python
for year in FCF_proj.index:
    FCF_proj.loc[year,"Free Cash Flow"] = last_year["Free Cash Flow"] * ((1 + company_growth_rate) ** (year-last_year.name))
```

```python
long_term_growth_rate = 0.06    # Indian Economy Growth Rate

# Calculate Residual Free Cash Flow
FCF_proj.loc["Residual","Free Cash Flow"] = FCF_proj.loc[FCF_proj.index[-1],"Free Cash Flow"] * (1 + long_term_growth_rate)/(CAPM_discount_rate - long_term_growth_rate)
```

## Calculate Present Value of FCF Projections

```python
for year in FCF_proj.index:
    if year=="Residual":
        year1 = 2029
    else:
        year1 = year
    FCF_proj.loc[year,"Present Value FCF"] = FCF_proj.loc[year,"Free Cash Flow"] / ((1+CAPM_discount_rate) ** (year1-last_year.name))
```

```python
#ASSUMPTIONS

n = 5
revenue_growth_T = 0.02
ebit_perc_T = 0.16
tax_perc_T = 0.24
dna_perc_T = 0.04
capex_perc_T = 0.17
nwc_perc_T = 0.05
TGR = 0.06  # Terminal Growth Rate taken = India's Economic Growth Rate
```

```python
def interpolate(initial_value, terminal_value, nyears):
    return np.linspace(initial_value, terminal_value, nyears)
```

```python
years = range(df.index[-1]+1, df.index[-1] + n + 1)
df_proj = pd.DataFrame(index=years, columns=df.columns)
# linear interpolation
df_proj["Revenue growth"] = interpolate(last_year["Revenue growth"], revenue_growth_T, n) 
df_proj["EBIT of Sales"] = interpolate(last_year["EBIT of Sales"], ebit_perc_T, n) 
df_proj["DnA of Sales"] = interpolate(last_year["DnA of Sales"], dna_perc_T, n) 
df_proj["CapEx of Sales"] = interpolate(last_year["CapEx of Sales"], capex_perc_T, n) 
df_proj["Tax of EBIT"] = interpolate(last_year["Tax of EBIT"], tax_perc_T, n) 
df_proj["Working Capital of Sales"] = interpolate(last_year["Working Capital of Sales"], nwc_perc_T, n) 

# cumulative values
df_proj["Total Revenue"] = last_year["Total Revenue"] *(1+df_proj["Revenue growth"]).cumprod() 
df_proj["EBIT"] = last_year["EBIT"] *(1+df_proj["EBIT of Sales"]).cumprod() 
df_proj["Capital Expenditure"] = last_year["Capital Expenditure"] *(1+df_proj["CapEx of Sales"]).cumprod() 
df_proj["Depreciation And Amortization"] = last_year["Depreciation And Amortization"] *(1+df_proj["DnA of Sales"]).cumprod() 
df_proj["Delta Working Capital"] = last_year["Delta Working Capital"] *(1+df_proj["Working Capital of Sales"]).cumprod() 
df_proj["Tax Provision"] = last_year["Tax Provision"] *(1+df_proj["Tax of EBIT"]).cumprod() 
df_proj["EBIAT"] = df_proj["EBIT"] - df_proj["Tax Provision"]
```

```python
df_proj["Free Cash Flow"] = df_proj["EBIAT"] + df_proj["Depreciation And Amortization"] - df_proj["Capital Expenditure"] - df_proj["Delta Working Capital"]
```
